<H1>League Totals</H1>
<p>Displays bar graph of stat per team</p>

```python
#Grab the league info

#!pip install matplotlib
#!pip install espn_api
from espn_api.basketball import League
import matplotlib.pyplot as plt
league = League(league_id=509828671, year=2024, espn_s2='ESPN_S2',swid='{SWID}')
teams = league.teams
```

```python
#Initiate league dictionary for total
team_stat_totals = {}
for team in league.teams:
    team_stat_totals[team.team_name] = {'PTS': 0, 'BLK': 0, 'STL': 0, 'AST': 0, 'REB': 0, 'TO': 0, 'FGM': 0, 'FGA': 0, 'FTM': 0, 'FTA': 0, '3PTM': 0, 'FG%': 0, 'FT%': 0}
```

```python
#Populate the league dictionary
for week in range(0,7):
    box = league.box_scores(week)
    for score in box:
        # Home team stats
        home_team_name = score.home_team.team_name
        for stat in team_stat_totals[home_team_name]:
            team_stat_totals[home_team_name][stat] += score.home_stats[stat]['value']
        # Away team stats
        away_team_name = score.away_team.team_name
        for stat in team_stat_totals[away_team_name]:
            team_stat_totals[away_team_name][stat] += score.home_stats[stat]['value']
            
#Correcting the percentages
for team in team_stat_totals.keys():
    team_stat_totals[team]["FG%"] = team_stat_totals[team]["FGM"]/team_stat_totals[team]["FGA"]
    team_stat_totals[team]["FT%"] = team_stat_totals[team]["FTM"]/team_stat_totals[team]["FTA"]
```

```python
#Plot stat function
def plot_stat(stat):
    team_names = ["Team 1","Team 2","Team 3","Team 4","Team 5","Team 6","Team 7","Team 8"]
    team_stats = []
    for key in team_stat_totals.keys():
        team_stats.append(team_stat_totals[key][stat])
    
    plt.figure(figsize=(10, 6))
    plt.bar(team_names, team_stats, color='blue')
    plt.xlabel('Teams')
    plt.ylabel(stat)
    plt.title(f'{stat} by Each Team')
    plt.xticks(rotation=45)
    plt.show()
```

```python
stats_to_plot = ['PTS', 'BLK', 'STL', 'AST', 'REB', 'TO', 'FGM', 'FGA', 'FTM', 'FTA', '3PTM', 'FG%', 'FT%']
for stat in stats_to_plot:
    plot_stat(stat)
```

```python
plot_stat("PTS")
```

```python
import pandas as pd

# Assuming 'team_stat_totals' is your dictionary
df = pd.DataFrame.from_dict(team_stat_totals, orient='index')
df.to_excel('team_stat_totals.xlsx', sheet_name='Stats')
```
